```python
a = 26600 #km
i = 1.10654 #rad
e = 0.74
omega = 5*np.pi/180 #rad
Omega = 90*np.pi/180 #rad \\ RA
M0 = 10*np.pi/180 #rad
mu = 3.986 * 10**5 #km^3/s^2
period = 2*np.pi*np.sqrt(a**3/mu)
h = np.sqrt(mu*a*(1-e**2))


theta = 40*np.pi/180 #guessing this doesn't matter bc it's only where vehicle is along orbit // TA
J2 = 1082.63e-6
R = 6378 #km
u = omega + theta
r = (h**2/mu)*(1/(1+(e*np.cos(theta))))

#using Gauss' Variational Equations
# y0 = np.array([[h],
#               [e],
#               [theta],
#               [Omega],
#               [i],
#               [omega]])

y0 = np.array([h,e,theta,Omega,i,omega]).T

# h_dot =  - (3/2) * ((J2*mu*R**2)/(r**3)) * (np.sin(i))**2 * np.sin(2*u)
# e_dot = (3/2) * ((J2*mu*R**2)/(h*r**3)) * (h**2/(mu*r)*np.sin(theta)*(3*(np.sin(i))**2*(np.sin(u))**2 -1) - (np.sin(2*u)*(np.sin(i)**2)*((3+(e*np.cos(theta)))*np.cos(theta) + e)))
# theta_dot = (h/(r**2)) + ((3/2) * ((J2*mu*R**2)/(e*h*r**3))) * ((h**2/(mu*r))*np.cos(theta)*(3*(np.sin(i))**2*(np.sin(u))**2 -1) + ((2 + (e*np.cos(theta)))*np.sin(2*u)*(np.sin(i))**2*np.sin(theta)))
# Omega_dot = -3 * ((J2*mu*R**2)/(h*r**3)) * (np.sin(u)**2) * np.cos(i)
# i_dot = -(3/4) * ((J2*mu*R**2)/(h*r**3)) * np.sin(2*u) * np.sin(2*i)
# omega_dot = (3/2) * ((J2*mu*R**2)/(e*h*r**3)) * ((h**2/(mu*r))*np.cos(theta)*(1 - (3*(np.sin(i))**2*(np.sin(u))**2)) - ((2 + (e*np.cos(theta)))*np.sin(2*u)*(np.sin(i))**2*np.sin(theta) + (2*e*(np.cos(i)**2)*(np.sin(u)**2))))

# f = np.array([[h_dot],
#               [e_dot],
#               [theta_dot],
#               [Omega_dot],
#               [i_dot],
#               [omega_dot]])

# f = np.array([h_dot, e_dot, theta_dot, Omega_dot, i_dot, omega_dot]).T

def rates(t,f):
    h = f[0]
    e = f[1]
    theta = f[2]
    Omega = f[3]
    i = f[4]
    omega = f[5]

    u = omega + theta
    r = (h**2/mu)*(1/(1+(e*np.cos(theta))))

    h_dot =  - (3/2) * ((J2*mu*R**2)/(r**3)) * (np.sin(i)**2) * np.sin(2*u)
    e_dot = (3/2) * ((J2*mu*R**2)/(h*r**3)) * (h**2/(mu*r)*np.sin(theta)*(3*(np.sin(i))**2*(np.sin(u))**2 -1) - (np.sin(2*u)*(np.sin(i)**2)*((3+(e*np.cos(theta)))*np.cos(theta) + e)))
    theta_dot = (h/(r**2)) + ((3/2) * ((J2*mu*R**2)/(e*h*r**3))) * ((h**2/(mu*r))*np.cos(theta)*(3*(np.sin(i))**2*(np.sin(u)**2) -1) + ((2 + (e*np.cos(theta)))*np.sin(2*u)*(np.sin(i)**2)*np.sin(theta)))
    Omega_dot = -3 * ((J2*mu*R**2)/(h*r**3)) * (np.sin(u)**2) * np.cos(i)
    i_dot = -(3/4) * ((J2*mu*R**2)/(h*r**3)) * np.sin(2*u) * np.sin(2*i)
    omega_dot = (3/2) * ((J2*mu*R**2)/(e*h*r**3)) * ((h**2/(mu*r))*np.cos(theta)*(1 - (3*(np.sin(i)**2)*(np.sin(u)**2))) - ((2 + (e*np.cos(theta)))*np.sin(2*u)*(np.sin(i)**2)*np.sin(theta) + (2*e*(np.cos(i)**2)*(np.sin(u)**2))))

    # dfdt = np.array([[h_dot],
    #           [e_dot],
    #           [theta_dot],
    #           [Omega_dot],
    #           [i_dot],
    #           [omega_dot]])

    dfdt = np.array([h_dot, e_dot, theta_dot, Omega_dot, i_dot, omega_dot]).T
    
    return dfdt

sol = scipy.integrate.solve_ivp(rates, [0, 100*24*(60**2)], y0, method='RK45', dense_output=True, max_step=0.1, rtol=1e-8)
```
