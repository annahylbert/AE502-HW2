```python
#givens
J2 = 0.00108
R = 6370 #km
GM = 3.986 * 10**5 #km^3/s^2
r_p = 600 #km
period = 8 * 60 * 60 #s // to orbit earth three times in one day period must be 8 hrs
n = 2 * np.pi / period #mean motion in rad/s


#in order for omega dot to be zero, i must be such that cos(i)^2*5 - 1 = 0
i = np.arccos(np.sqrt(0.2))
print('i (rad) =', i)
print('i (deg) = ', i*180/np.pi)

#in order for r_a to be maximized e and a should also be maximized through the following equation r_a = a*(1+e) 
#a and e are also constrained through the following condition r_p = a*(1-e) where a = r_p/(1-e) so we want e to be as close to 1 as possible 

a = ((period**2 * GM) / (4*(np.pi**2))) ** (1/3)
e = 1 - ((r_p+R)/a)

print('a (km) =', a)
print('e =', e)

Omega_dot = -1.5 * n * J2 * (R/a)**2 * (np.cos(i)/((1-e**2)**2))
print("Lowest Omega_dot Drift Rate =", Omega_dot)

```

**Problem 3**

```python
a = 26600 #km
i0 = 1.10654 #rad
e0 = 0.74
omega0 = 5*np.pi/180 #rad
Omega0 = 90*np.pi/180 #rad \\ RA
M0 = 10*np.pi/180 #rad
mu = 398600 #km^3/s^2
period = 2*np.pi*np.sqrt(a**3/mu)
h0 = np.sqrt(mu*a*(1-e0**2))


theta0 = 40*np.pi/180 # TA
J2 = 1082.63e-6
R = 6378 #km
u = omega0 + theta0
r = (h0**2/mu)*(1/(1+(e0*np.cos(theta0))))


y0 = np.array([h0,e0,theta0,Omega0,i0,omega0]).T

def rates(t,f):
    h = f[0]
    e = f[1]
    theta = f[2]
    Omega = f[3]
    i = f[4]
    omega = f[5]

    u = omega + theta
    r = a*(1-e**2)/(1+(e*np.cos(theta)))

    h_dot =  - (3/2) * ((J2*mu*R**2)/(r**3)) * (np.sin(i)**2) * np.sin(2*u)
    e_dot = (3/2) * ((J2*mu*(R**2))/(h*(r**3))) * (((h**2/(mu*r))*np.sin(theta)*((3*(np.sin(i)**2)*(np.sin(u)**2)) -1)) - (np.sin(2*u)*(np.sin(i)**2)*((2+(e*np.cos(theta)))*np.cos(theta) + e)))
    theta_dot = (h/(r**2)) + ((3/2) * ((J2*mu*R**2)/(e*h*r**3))) * ((h**2/(mu*r))*np.cos(theta)*(3*(np.sin(i))**2*(np.sin(u)**2) -1) + ((2 + (e*np.cos(theta)))*np.sin(2*u)*(np.sin(i)**2)*np.sin(theta)))
    Omega_dot = -3 * ((J2*mu*R**2)/(h*r**3)) * (np.sin(u)**2) * np.cos(i)
    i_dot = -(3/4) * ((J2*mu*R**2)/(h*r**3)) * np.sin(2*u) * np.sin(2*i)
    omega_dot = (3/2) * ((J2*mu*R**2)/(e*h*r**3)) * ((-h**2/(mu*r))*np.cos(theta)*((3*(np.sin(i)**2)*(np.sin(u)**2))-1) - ((2 + (e*np.cos(theta)))*np.sin(2*u)*(np.sin(i)**2)*np.sin(theta) - (2*e*(np.cos(i)**2)*(np.sin(u)**2))))

    dfdt = np.array([h_dot, e_dot, theta_dot, Omega_dot, i_dot, omega_dot]).T
    
    return dfdt

y0_curtis = np.array([55839, 0.17136, 40*np.pi/180, 45*np.pi/180, 28*np.pi/180, 30*np.pi/180]).T
t_f = 100*24*(60**2)
t_f_curtis = 2*24*(60**2)



sol = scipy.integrate.solve_ivp(rates, [0, t_f], y0, method='RK45', dense_output=True, first_step=period/1000, max_step=10, rtol=1e-14)
```

```python
plt.style.use('default')
fig = plt.figure(figsize=(15, 13), tight_layout=True)

t = np.linspace(0, t_f, np.shape(sol.y)[1])
Omega_array = sol.y[3]
omega_array = sol.y[5]
h_array = sol.y[0]
e_array = sol.y[1]
i_array = sol.y[4]

Omega_plot = fig.add_subplot(511)
Omega_plot.plot(t/(3600*24), (Omega_array-y0[3])*180/np.pi, linewidth=1.2, color='black')
Omega_plot.set_xlabel('Time (Days)')
Omega_plot.set_title('Right Ascension of the Ascending Node (Degrees)')

omega_plot = fig.add_subplot(512)
omega_plot.plot(t/(3600*24), (omega_array-y0[5])*180/np.pi, linewidth=1.2, color='black')
omega_plot.set_xlabel('Time (Days)')
omega_plot.set_title('Arguement of Perigee (Degrees)')

h_plot = fig.add_subplot(513)
h_plot.plot(t/(3600*24), h_array - y0[0], linewidth=1.2, color='black')
h_plot.set_xlabel('Time (Days)')
h_plot.set_title(r'Angular Momentum ($\frac{km^2}{s}$)')

e_plot = fig.add_subplot(514)
e_plot.plot(t/(3600*24), e_array - y0[1], linewidth=1.2, color='black')
e_plot.set_xlabel('Time (Days)')
e_plot.set_title('Eccentricity')

i_plot = fig.add_subplot(515)
i_plot.plot(t/(3600*24), (i_array-y0[4])*180/np.pi, linewidth=1.2, color='black')
i_plot.set_xlabel('Time (Days)')
i_plot.set_title('Inclination (Degrees)')

fig.savefig('Perturbed_orbital_elements.png')
```

The inclination, eccentricy, and angular momentum all oscillate but have the same average averaged value over the 100 day period. However, the right ascension of the ascending node and arguement of perigee both oscillate and grow/decay linearly over the one hundred day period. Because the oscillating portion of these graphs average out over time, this is why they are sometimes averaged to only account for the linear growth/decay.
